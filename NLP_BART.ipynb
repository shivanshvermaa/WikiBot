```python
# Print first article
print("--- Full Article ---")
print(df["Article"][0])
print(" ")
print("--- Summary ---")
print(df["Summary"][0])
```

```python
# Define the categories you want to select
categories = ["People", "Arts", "History","Technology","Mathematics" ,"Biology_and_health_sciences","Physical_sciences" ,"Everyday_life", "Geography", "Philosophy_and_religion", "Society_and_social_sciences"]

# Create an empty list to store the selected rows
selected_rows = []

# Loop through each category and select 10 rows
for category in categories:
    category_rows = df[df["Category"] == category].head(25)
    selected_rows.append(category_rows)

# Concatenate all the selected rows into a new dataframe
df = pd.concat(selected_rows)
```

```python
df_sample = df
df_sample
```

```python
from transformers import BartForConditionalGeneration, BartTokenizer

# Load the pre-trained BART model and tokenizer
model = BartForConditionalGeneration.from_pretrained('facebook/bart-large')
tokenizer = BartTokenizer.from_pretrained('facebook/bart-large')
```

```python
input_text = df_sample['Article_clean'].tolist()

# Tokenize the input text
tokenized_input = tokenizer.batch_encode_plus(input_text, padding = True, truncation = True, max_length = 512, return_tensors = 'pt')
```

```python
# Generate summaries using the pre-trained BART model
summaries = model.generate(tokenized_input['input_ids'], 
                            attention_mask = tokenized_input['attention_mask'],
                            max_length = 200, min_length = 10, num_beams =  4)
```

```python
# Decoding the generated summaries
decoded_summaries = [tokenizer.decode(g, skip_special_tokens = True, clean_up_tokenization_spaces = False) for g in summaries]
```

```python
# Print the generated summary
print("--- Generated Summary ---")
print(decoded_summaries[0])
print(" ")
```
